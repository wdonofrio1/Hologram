What is the cost of all data used in the linked data?

What do you mean by the "linked data"?  If the rt_amt is the cost, you can join this to the sim_card_plan_history_df on bundle_id.  However, it appears that this table contains slowly changing dimiensions, so you would have to also add a between to join on dates.  You then have to join assest_id to the profile_installation_df and the p_id to the usage_events_df.

```python
import pandas as pd

FAR_FUTURE = pd.Timestamp('2099-12-31')

# --- Prep: rename the date-window columns so each table's window is unambiguous,
#           uppercase the tech codes so lte/LTE etc. match, and fill open-ended
#           end dates so "still active" rows participate in the joins.
rates = (rate_card_df
    .rename(columns={'beg_dttm': 'rate_beg', 'end_dttm': 'rate_end'})
    .assign(tech_cd=lambda d: d['tech_cd'].str.upper(),
            rate_end=lambda d: d['rate_end'].fillna(FAR_FUTURE)))

plans = (sim_card_plan_history_df
    .rename(columns={'eff_dttm': 'plan_beg', 'x_dttm': 'plan_end'})
    .assign(plan_end=lambda d: d['plan_end'].fillna(FAR_FUTURE)))

profiles = (profile_installation_df
    .rename(columns={'beg_dttm': 'inst_beg', 'end_dttm': 'inst_end'})
    .assign(inst_end=lambda d: d['inst_end'].fillna(FAR_FUTURE)))

events = usage_events_df.assign(tech=lambda d: d['tech'].str.upper())

# --- Step 1: rate_card x sim_card_plan_history on bundle_id with overlapping windows
rate_plan = (rates
    .merge(plans, on='bundle_id', how='inner')
    .query('rate_beg <= plan_end and rate_end >= plan_beg'))

# --- Step 2: x profile_installation on asset_id with overlapping windows
rate_plan_inst = (rate_plan
    .merge(profiles, on='asset_id', how='inner')
    .query('plan_beg <= inst_end and plan_end >= inst_beg'))

# --- Step 3: x usage_events on pid (and cc1/cc2/tech so each event picks its correct rate)
#             and require evt_dttm to fall inside all three active windows
joined = rate_plan_inst.merge(
    events,
    left_on=['pid', 'cc1', 'cc2', 'tech_cd'],
    right_on=['pid', 'cc1', 'cc2', 'tech'],
    how='inner',
)

results = (joined
    .query('evt_dttm >= rate_beg and evt_dttm < rate_end')
    .query('evt_dttm >= plan_beg and evt_dttm < plan_end')
    .query('evt_dttm >= inst_beg and evt_dttm < inst_end')
    .assign(cost=lambda d: d['mb'] * d['rt_amt']))

# Total
print(f"Total cost: ${results['cost'].sum():,.4f}")
```
